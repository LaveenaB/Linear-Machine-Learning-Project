### Data Preprocessing

#### mapping the tier no. to colleges


```python
college_tier_mapping = {College:1 for College in college['Tier 1']}
college_tier_mapping.update({College:2 for College in college['Tier 2']})
college_tier_mapping.update({College:3 for College in college['Tier 3']})
```

```python
college_tier_mapping
```

#### adding tier mapping to our candidates dataframe with new column named as tier_college


```python
candidates['tier_college'] = candidates['College'].map(college_tier_mapping)
```

```python
candidates.head()
```

#### mapping the cities being metro city as 1 and non-metro city as 0


```python
city_mapping = {City:0 for City in cities['non-metro cities']}
city_mapping.update({City:1 for City in cities['Metrio City']})
```

```python
city_mapping
```

#### Adding new column (Metro city) to candidates dataframe denoting metro as 1 and non-metro as 0

```python
candidates['metro_city'] = candidates['City'].map(city_mapping)
```

```python
# checking again for any null values
candidates.isnull().sum()
```

## Model 3: XGB Boosting

### Importing Libraries

```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

### Standardizing the data 

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
param_grid = {
    'n_estimators': [100,300],
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}
```

```python
xgb_reg = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

grid_search = GridSearchCV(estimator=xgb_reg, param_grid=param_grid, cv=3)

grid_search.fit(X_train_scaled, y_train)
```

```python
y_test_pred_xgb = grid_search.predict(X_test_scaled)
```
